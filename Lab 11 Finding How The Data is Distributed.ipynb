### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

### Step 2: Examine the Structure of the Data


- Display the column names, data types, and summary information to understand the data structure.

- Objective: Gain insights into the dataset's shape and available variables.


```python
## Write your code here
# Column names and data types
df.info()

# Summary statistics
df.describe(include='all')
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
## Write your code here
# Total missing values per column
missing_values = df.isnull().sum()
print("Missing values:\n", missing_values)

# Drop rows with missing values in important columns
df_cleaned = df.dropna(subset=['Employment', 'JobSat', 'YearsCodePro'])

# Reset index after dropping rows
df_cleaned.reset_index(drop=True, inplace=True)
```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
## Write your code here
# Value counts
print("Employment:\n", df_cleaned['Employment'].value_counts(), "\n")
print("Job Satisfaction:\n", df_cleaned['JobSat'].value_counts(), "\n")
print("Years of Pro Coding Experience:\n", df_cleaned['YearsCodePro'].value_counts().sort_index())
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
## Write your code here
# JobSat distribution as pie chart
job_sat_counts = df_cleaned['JobSat'].value_counts()
job_sat_counts.plot.pie(autopct='%1.1f%%', figsize=(8, 8), title='Job Satisfaction Distribution')
plt.ylabel('')
plt.show()
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here
from collections import Counter

# Split and count languages
def get_lang_counts(column):
    return Counter(lang for row in df_cleaned[column].dropna() for lang in row.split(';'))

worked_with = get_lang_counts('LanguageHaveWorkedWith')
want_to_work_with = get_lang_counts('LanguageWantToWorkWith')

# Top 10 from each
top_worked = dict(Counter(worked_with).most_common(10))
top_want = dict(Counter(want_to_work_with).most_common(10))

# Bar chart comparison
df_langs = pd.DataFrame({
    'Worked With': pd.Series(top_worked),
    'Want To Work With': pd.Series(top_want)
}).fillna(0)

df_langs.plot(kind='bar', figsize=(12,6), title='Programming Language Preferences')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here
# Clean 'Country' and 'RemoteWork'
df_cleaned['Country'] = df_cleaned['Country'].fillna('Unknown')
remote_region = pd.crosstab(df_cleaned['Country'], df_cleaned['RemoteWork'])

# Show only top 10 countries
top_countries = df_cleaned['Country'].value_counts().head(10).index
remote_region_top = remote_region.loc[top_countries]

# Plot heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(remote_region_top, annot=True, cmap="Blues", fmt="d")
plt.title('Remote Work Distribution by Country (Top 10)')
plt.ylabel('Country')
plt.xlabel('Remote Work Status')
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
!pip install scipy
```

```python
from scipy.stats import spearmanr

# Convert experience to numeric safely
df_cleaned.loc[:, 'YearsCodePro'] = pd.to_numeric(df_cleaned['YearsCodePro'], errors='coerce')

# Map job satisfaction to numeric safely
job_sat_map = {
    'Very satisfied': 5,
    'Slightly satisfied': 4,
    'Neither satisfied nor dissatisfied': 3,
    'Slightly dissatisfied': 2,
    'Very dissatisfied': 1
}
df_cleaned.loc[:, 'JobSatScore'] = df_cleaned['JobSat'].map(job_sat_map)

# Drop rows with missing values in the two columns
corr_data = df_cleaned[['YearsCodePro', 'JobSatScore']].dropna()

# Calculate Spearman correlation
corr, _ = spearmanr(corr_data['YearsCodePro'], corr_data['JobSatScore'])
print(f"Spearman correlation between experience and job satisfaction: {corr:.2f}")
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
## Write your code here
edu_emp = pd.crosstab(df_cleaned['EdLevel'], df_cleaned['Employment'])

# Stacked bar plot
edu_emp.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Employment Type by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```
