### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
## Write your code here
age_order = [
     'Under 18 years old', '18-24 years old', '25-34 years old',
     '35-44 years old', '45-54 years old', '55-64 years old', '65 years or older']
print("\n--- Generating: Line Chart - Median Compensation by Age Group ---")

# --- 1. Define Column Names ---
age_col = 'Age'
comp_col = 'ConvertedCompYearly'

# --- 2. Check if Required Columns Exist ---
if age_col in df.columns and comp_col in df.columns:

    # --- 3. Prepare Data ---
    data_comp_age = df[[age_col, comp_col]].copy()
    data_comp_age[comp_col] = pd.to_numeric(data_comp_age[comp_col], errors='coerce')
    data_comp_age = data_comp_age[data_comp_age[age_col].isin(age_order)]
    data_comp_age[age_col] = pd.Categorical(data_comp_age[age_col],
                                            categories=age_order,
                                            ordered=True)
    data_comp_age.dropna(subset=[age_col, comp_col], inplace=True)
    print(f"Using {len(data_comp_age)} rows with valid Age and Compensation.")

    # --- 4. Calculate Median Compensation per Age Group ---
    median_compensation = data_comp_age.groupby(age_col, observed=False)[comp_col].median()
    median_compensation = median_compensation.reindex(age_order)
    print("\nMedian Yearly Compensation by Age Group:")
    print(median_compensation)

    # --- 5. Check if Data is Available for Plotting ---
    if median_compensation.isnull().all():
        print("Error: No valid median compensation data calculated. Cannot generate plot.")
    elif median_compensation.dropna().empty:
        print("Warning: All median compensation values are NaN. Plot will be empty.")

    # --- 6. Create the Line Chart ---
    if not median_compensation.dropna().empty:
        plt.figure(figsize=(12, 7))
        plt.plot(median_compensation.index, median_compensation.values,
                 marker='o', linestyle='-', color='royalblue')

        # --- 7. Customize the Plot ---
        plt.title('Median Yearly Compensation by Age Group', fontsize=16)
        plt.xlabel('Age Group', fontsize=12)
        plt.ylabel('Median Yearly Compensation (USD)', fontsize=12)
        plt.xticks(rotation=45, ha='right')

        # --- Format y-axis using FuncFormatter (CORRECTED PART) ---
        ax = plt.gca() # Get current axes
        # Define a function to format the tick value
        def currency_formatter(x, pos):
            # x is the value, pos is the position (often ignored)
            if pd.isna(x): # Handle potential NaN values if they reach the formatter
                 return ''
            return '${:,.0f}'.format(x) # Use Python's format() method

        # Create the formatter object from the function
        formatter = mticker.FuncFormatter(currency_formatter)
        # Apply the formatter to the y-axis
        ax.yaxis.set_major_formatter(formatter)
        # --- End of Correction ---

        plt.grid(True, which='major', linestyle='--', linewidth=0.5, alpha=0.7)
        plt.tight_layout() # This should work now
        plt.show()
    else:
        pass # Message already printed

else:
    print(f"Line chart skipped: Required columns '{age_col}' or '{comp_col}' not found.")
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python
import numpy as np
```

```python
## Write your code here
print("\n--- Generating: Line Chart - Median Compensation for Ages 25-45 ---")

# --- 1. Define Column Names & Target Age Groups ---
age_col = 'Age'
comp_col = 'ConvertedCompYearly'
# Identify the age group strings that cover the 25-45 range
# Note: '45-54 years old' includes age 45.
target_age_groups = ['25-34 years old', '35-44 years old', '45-54 years old']

# --- 2. Check if Required Columns Exist ---
if age_col in df.columns and comp_col in df.columns:

    # --- 3. Prepare Data (Filter first) ---
    # Select relevant columns and copy
    data_comp_age_subset = df[[age_col, comp_col]].copy()

    # Ensure 'ConvertedCompYearly' is numeric
    data_comp_age_subset[comp_col] = pd.to_numeric(data_comp_age_subset[comp_col], errors='coerce')

    # **Filter for the target age groups**
    data_comp_age_subset = data_comp_age_subset[data_comp_age_subset[age_col].isin(target_age_groups)]

    # Ensure 'Age' is categorical with the correct order *for the subset*
    data_comp_age_subset[age_col] = pd.Categorical(data_comp_age_subset[age_col],
                                                   categories=target_age_groups, # Use the subset order
                                                   ordered=True)

    # Drop rows where either Age (within the subset) or Compensation is NaN
    data_comp_age_subset.dropna(subset=[age_col, comp_col], inplace=True)
    print(f"Using {len(data_comp_age_subset)} rows from age groups {target_age_groups} with valid Compensation.")

    # --- 4. Calculate Median Compensation for the Subset ---
    # Group by the ordered Age category (subset) and calculate the median
    # observed=False ensures all categories in target_age_groups appear if specified in Categorical
    median_compensation_subset = data_comp_age_subset.groupby(age_col, observed=False)[comp_col].median()

    # Reindex explicitly to be sure of the order (optional but safe)
    median_compensation_subset = median_compensation_subset.reindex(target_age_groups)

    print("\nMedian Yearly Compensation for Age Groups 25-54:")
    print(median_compensation_subset)

    # --- 5. Check if Data is Available for Plotting ---
    if median_compensation_subset.isnull().all():
        print("Error: No valid median compensation data calculated for the subset. Cannot generate plot.")
    elif median_compensation_subset.dropna().empty:
        print("Warning: All median compensation values for the subset are NaN. Plot will be empty.")

    # --- 6. Create the Line Chart ---
    if not median_compensation_subset.dropna().empty:
        plt.figure(figsize=(10, 6)) # Adjust size slightly if needed
        plt.plot(median_compensation_subset.index, median_compensation_subset.values,
                 marker='o', linestyle='-', color='forestgreen') # Different color

        # --- 7. Customize the Plot ---
        plt.title('Median Yearly Compensation for Ages 25-45', fontsize=16) # Update title
        plt.xlabel('Age Group', fontsize=12)
        plt.ylabel('Median Yearly Compensation (USD)', fontsize=12)

        # No need to rotate x-axis labels for only 3 categories
        plt.xticks(rotation=0)

        # --- Format y-axis using FuncFormatter (reuse from previous step) ---
        ax = plt.gca()
        def currency_formatter(x, pos):
             if pd.isna(x):
                 return ''
             return '${:,.0f}'.format(x)
        formatter = mticker.FuncFormatter(currency_formatter)
        ax.yaxis.set_major_formatter(formatter)
        # --- End of Formatting ---

        plt.grid(True, which='major', linestyle='--', linewidth=0.5, alpha=0.7)
        plt.tight_layout()
        plt.show()
    else:
        pass # Message already printed

else:
    print(f"Line chart skipped: Required columns '{age_col}' or '{comp_col}' not found.")
```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python
## Write your code here
print("\n--- Generating: Line Chart - Median Job Satisfaction (Points 6) by Years of Professional Experience ---")

# --- 1. Define Column Names ---
exp_col = 'YearsCodePro'
sat_col = 'JobSatPoints_6'

# --- 2. Check if Required Columns Exist ---
if exp_col in df.columns and sat_col in df.columns:

    # --- 3. Prepare Data ---
    # Select relevant columns and copy
    data_exp_sat = df[[exp_col, sat_col]].copy()

    # --- Clean 'YearsCodePro' column ---
    # Map string values to numeric representations
    data_exp_sat[exp_col] = data_exp_sat[exp_col].replace('Less than 1 year', 0)
    data_exp_sat[exp_col] = data_exp_sat[exp_col].replace('More than 50 years', 51) # Use 51 to represent >50

    # Convert the column to numeric, coercing errors (handles remaining strings/potential issues)
    data_exp_sat[exp_col] = pd.to_numeric(data_exp_sat[exp_col], errors='coerce')

    # --- Clean 'JobSatPoints_6' column ---
    data_exp_sat[sat_col] = pd.to_numeric(data_exp_sat[sat_col], errors='coerce')

    # --- Drop rows with missing values in either column ---
    data_exp_sat.dropna(subset=[exp_col, sat_col], inplace=True)
    # Convert experience column to integer after handling NaNs for cleaner grouping/indexing
    data_exp_sat[exp_col] = data_exp_sat[exp_col].astype(int)

    print(f"Using {len(data_exp_sat)} rows with valid numeric Experience and Job Satisfaction.")

    # --- 4. Calculate Median Job Satisfaction per Year of Experience ---
    # Group by the numeric years of professional experience and calculate the median satisfaction
    median_satisfaction_by_exp = data_exp_sat.groupby(exp_col)[sat_col].median()

    # Sort by index (YearsCodePro) to ensure the line chart is chronological
    median_satisfaction_by_exp = median_satisfaction_by_exp.sort_index()

    print(f"\nMedian Job Satisfaction ({sat_col}) by Years of Professional Experience ({exp_col}):")
    # Print head and tail to avoid excessively long output if many years exist
    print(median_satisfaction_by_exp.head())
    print("...")
    print(median_satisfaction_by_exp.tail())


    # --- 5. Check if Data is Available for Plotting ---
    if median_satisfaction_by_exp.empty:
        print("Error: No valid median satisfaction data calculated after grouping by experience. Cannot generate plot.")
    elif median_satisfaction_by_exp.isnull().all():
         print("Warning: All median satisfaction values are NaN after grouping. Plot might be empty or incorrect.")

    # --- 6. Create the Line Chart ---
    if not median_satisfaction_by_exp.empty:
        plt.figure(figsize=(14, 7)) # Wider figure for potentially many years
        plt.plot(median_satisfaction_by_exp.index, median_satisfaction_by_exp.values,
                 marker='.', linestyle='-', color='purple') # Use dots as markers

        # --- 7. Customize the Plot ---
        plt.title(f'Median Job Satisfaction ({sat_col}) by Years of Professional Experience', fontsize=16)
        plt.xlabel('Years of Professional Coding Experience', fontsize=12)
        plt.ylabel(f'Median Job Satisfaction ({sat_col})', fontsize=12)

        # Set x-axis limits and ticks for better readability if needed
        # Example: Show ticks every 5 years up to the max found (or 51)
        max_exp = median_satisfaction_by_exp.index.max()
        plt.xticks(np.arange(0, max_exp + 1, 5)) # Ticks every 5 years starting from 0
        plt.xlim(left=-1, right=max_exp + 1) # Add slight padding to x-axis limits


        plt.grid(True, which='major', linestyle='--', linewidth=0.5, alpha=0.7)
        plt.tight_layout()
        plt.show()
    else:
        pass # Message already printed

else:
    print(f"Line chart skipped: Required columns '{exp_col}' or '{sat_col}' not found.")
```
